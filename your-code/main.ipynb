#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
# Filtered users DataFrame
users_filtered = users[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]

# Filtered posts DataFrame
posts_filtered = posts[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']]
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merged_df = pd.merge(posts_filtered, users_filtered, on='userId', how='inner')
```

```python
#If the missing values are in key columns like userId, it likely means the post has no known author. In that case, it's best to drop those rows because we can't match them to a user.
#If the missing values are in numeric columns like CommentCount or ViewCount, it's safe to fill them with 0, assuming missing means no comments or no views recorded.

# Drop rows where userId is missing (can't merge them with user data)
merged_df = merged_df.dropna(subset=['userId'])

# Fill missing numeric columns with 0
merged_df[['CommentCount', 'ViewCount']] = merged_df[['CommentCount', 'ViewCount']].fillna(0)
```

```python
# Define the columns to check
columns_to_check = ['Score', 'ViewCount', 'CommentCount', 'Reputation', 'Views', 'UpVotes', 'DownVotes']

# Create an empty DataFrame for outliers
outliers_df = pd.DataFrame()

# Loop through each column to find outliers based on IQR
for col in columns_to_check:
    Q1 = merged_df[col].quantile(0.25)
    Q3 = merged_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Append outliers for this column
    outliers = merged_df[(merged_df[col] < lower_bound) | (merged_df[col] > upper_bound)]
    outliers_df = pd.concat([outliers_df, outliers])

# Drop duplicate rows (in case the same row is an outlier in multiple columns)
outliers_df = outliers_df.drop_duplicates()

# Export to CSV
outliers_df.to_csv("outliers.csv", index=False)
```
